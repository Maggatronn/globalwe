```python
df = pd.read_csv ('./original_codes.csv')
updated = pd.read_csv ('./catalog_codes.csv')
df[0:30]
```

```python
topics_dict = {}

for i in range(0,len(df)): 
    theme = df["code"][i]
    parent = df["parent code"][i]
    if parent not in topics_dict.keys() and isinstance(parent, str):
        topics_dict[parent] = [theme]
    if isinstance(theme, str) and isinstance(parent, str) and theme not in topics_dict[parent]:
        topics_dict[parent].append(theme)
topics_dict        
```

```python
parent_child_list = []

for i in range(0,len(df)): 
    theme = df["code"][i]
    parent = df["parent code"][i]    
    if (theme, parent) not in parent_child_list and isinstance(parent,str):
        parent_child_list.append((theme, parent))
        
parent_child_list        
```

```python
parent_list = []
child_list = []
parent_dict = []
child_dict = []

highlight_list = []
links_dict= []

parent_count = 0
child_count = 0


past_highlight = 0
child_tag_list = []
parent_tag_list = []
highlights_tags_dict = {}               

# creating highlights_tags_dict -- meaning each entry ius a highlight with a list of parent 
# and child tags associated, along with words and speaker

no_list = ["Quietest Voices", "Standout Highlight", "Climate Care", "Empowerment"]
cut_tag = "Delete this Highlight"

for num in range(0, len(df)):
#     if isinstance(df['code'][num], str):
    highlight = df['annotation_id'][num]
    speaker = df['speaker_name'][num]
    parent_tag = df['parent code'][num]
    child_tag = df['code'][num]    
    type_tag = df['code_type'][num]  
    if num > 0:
        past_highlight = df['annotation_id'][num-1]

    if highlight == past_highlight:
        if isinstance(child_tag, str) and child_tag not in child_tag_list and child_tag not in no_list:
            child_tag_list.append(child_tag)
        if isinstance(parent_tag, str) and parent_tag not in parent_tag_list and parent_tag not in no_list:
            parent_tag_list.append(parent_tag)
        if past_highlight not in list(highlights_tags_dict.keys()):
            highlights_tags_dict[past_highlight] = {
                "parent_tag_list": parent_tag_list,
                "child_tag_list": child_tag_list,
                "speaker": speaker,
                "type": type_tag
            }
    else:         
        child_tag_list = []
        parent_tag_list = []
        identity_tag = ''
        type_tag = ''
        identity_others_tag = ''
        if parent_tag not in parent_tag_list and isinstance(parent_tag, str) and parent_tag not in no_list: 
            parent_tag_list.append(parent_tag)
        if isinstance(child_tag, str) and child_tag not in child_tag_list and child_tag not in no_list:
            child_tag_list.append(child_tag)

len(highlights_tags_dict)
```

```python
themes_dict = {}
for highlight in highlights_tags_dict: 
    for tag in highlights_tags_dict[highlight]['child_tag_list']:
        for i in topics_dict.keys():
            if isinstance(i, str):
                if tag in topics_dict[i] and tag not in topics_dict.keys():
                        parent = i
        if isinstance(tag, str):
            if tag not in list(themes_dict.keys()) and tag not in topics_dict.keys():
                themes_dict[tag] = {
                    'id': count,
                    'tag': tag,
                    'parent': parent,
                    'degree': 1,
                    'group': 2
                }
                count += 1
            if tag in list(themes_dict.keys()) and tag not in topics_dict.keys(): 
                themes_dict[tag]['degree'] += 1
themes_dict                
```
